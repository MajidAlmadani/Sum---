```python
# Define image preprocessing function
def preprocess_image(image_path):
    img = Image.open(image_path)
    img = img.resize((224, 224))  # Resize images to model's expected input size
    img = np.array(img) / 255.0   # Normalize pixel values to [0, 1]
    return img

# Load the dataset
def load_dataset(image_paths, labels):
    image_tensors = [preprocess_image(image) for image in image_paths]
    return np.array(image_tensors), np.array(labels)
```
